```python
from matplotlib import pyplot as plt
from collections import Counter
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
plt.plot(years,gdp,marker="o",color="pink",linestyle="dashed")
plt.ylabel("GDP")
plt.xlabel("YEARS")
plt.show()
```

```python
plt.plot(years,gdp,marker="o",color="pink",linestyle="dashed")
plt.ylabel("GDP")
plt.xlabel("YEARS")
plt.show()
plt.bar(range(len(years)),gdp)
#to give labeling on x-qaxis
plt.xticks(range(len(years)),years);
plt.xlabel("YEARS")
plt.ylabel("GDP")
plt.show()
```

```python
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]
array=[i//10*10 for i in grades]
print(array)
x=list(Counter(array).keys())
y=list(Counter(array).values())
plt.bar(range(len(x)),y) 
plt.xticks(range(len(x)),x)#x-axis
plt.yticks(range(len(y)),y)#y-axis
plt.show()
```

```python
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)
print(histogram)
```

```python
plt.bar([x + 5 for x in histogram.keys()], 
histogram.values(), 
10, edgecolor=(0, 0, 0))
```
